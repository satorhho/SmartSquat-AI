```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
import os
```

<h3>Augments Squat Data</h3>

 This function checks if the file is augmented data. If not it augments the data by flipping it. 
    <li>a0: vertically flips the image of the squat segment</li>
    <li>a1: horizontally flips the image</li>
    <li>a2: horizontally and vertiaclly flips the image</li>



```python
def augment_data(path, file_name, image):
    if not file_name[:1] == "a":
        cv.imwrite(f"{path}\\a0_{file_name}", cv.flip(image, 1))
        print(f"{path}\\{file_name}")
```

<h2>Extract and augments each data on the list</h2>


<strong>PARAMETERS AND FUNCTION FOR AUGMENTING DATA</strong>

```python
paths = [f"{os.path.abspath(os.getcwd())}\\train\\c_bot_1", f"{os.path.abspath(os.getcwd())}\\train\\c_mid_1",
        f"{os.path.abspath(os.getcwd())}\\train\\c_top_1", f"{os.path.abspath(os.getcwd())}\\train\\c_top_2",
        f"{os.path.abspath(os.getcwd())}\\train\\i_bot_1", f"{os.path.abspath(os.getcwd())}\\train\\i_mid_1",
        f"{os.path.abspath(os.getcwd())}\\train\\i_top_1", f"{os.path.abspath(os.getcwd())}\\train\\i_top_2",
        f"{os.path.abspath(os.getcwd())}\\validation\\c_bot_1", f"{os.path.abspath(os.getcwd())}\\validation\\c_mid_1",
        f"{os.path.abspath(os.getcwd())}\\validation\\c_top_1", f"{os.path.abspath(os.getcwd())}\\validation\\c_top_2",
        f"{os.path.abspath(os.getcwd())}\\validation\\i_bot_1", f"{os.path.abspath(os.getcwd())}\\validation\\i_mid_1",
        f"{os.path.abspath(os.getcwd())}\\validation\\i_top_1", f"{os.path.abspath(os.getcwd())}\\validation\\i_top_2",
        f"{os.path.abspath(os.getcwd())}\\test"] 
for path in paths:
    for root, directories, files in os.walk(path):
        for file in files:
            augment_data(path, file, cv.imread(f"{path}\\{file}"))
```
